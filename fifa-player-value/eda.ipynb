```python
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
submission = pd.read_csv(submission_path)
```

```python
# 데이터 타입 변환하여 메모리 사용량 줄이기
train[['age', 'stat_overall', 'stat_potential']] = train[['age', 'stat_overall', 'stat_potential']].astype(np.int32)
train[['reputation', 'stat_skill_moves']] = train[['reputation', 'stat_skill_moves']].astype(np.float32)
train.info()
```

## Feature Engineering

```python
# 계약기간 전처리(연도 추출)
train['contract_until'] = train['contract_until'].map(lambda x: x[-4:])
train['contract_until'].unique()
```

```python
# 남은 계약기간
train['contract_until'] = train['contract_until'].astype(np.int32)
train['contract_until'] = train['contract_until'] - 2018
```

```python
sns.countplot(train['contract_until'])
```

```python
# 성장 가능 스탯
train['stat_poten_left'] = train['stat_potential'] - train['stat_overall']
```

```python
# 평균 은퇴 연령까지 남은 나이
train['retire_age_left'] = 35 - train['age']
```
